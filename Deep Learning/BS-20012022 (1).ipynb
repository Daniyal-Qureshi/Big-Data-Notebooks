```
from tensorflow.keras.datasets import mnist
(train,trainclass),(test,testclass) = mnist.load_data()
train = train.reshape(60000,784)
test = test.reshape(10000,784)
train = train/255
train = train/255
train = train.astype('float32')
test = test.astype('float32')
```

```
trainclass = trainclass.astype('int')
```

```
import pandas as pd
pdtrain = pd.DataFrame(train)
pdclass = pd.DataFrame(trainclass, columns=['y'])
pddf = pd.concat([pdtrain,pdclass],axis=1)
```

```
pddf
```

```
df = spark.createDataFrame(pddf)
```

```
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.feature import OneHotEncoder
va = VectorAssembler(inputCols=[c for c in df.columns if c!='y'], outputCol='features')

df1 = va.transform(df)

ohe = OneHotEncoder(inputCols=['y'],
                            outputCols=['label'],
                            dropLast=False)
model = ohe.fit(df1)
train_df = model.transform(df1)
```

```
train_df.printSchema()
```

```
from tensorflow import keras
import tensorflow as tf
import horovod.spark.keras as hvd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation
from horovod.spark.common.store import DBFSLocalStore, Store
from tensorflow.keras.optimizers import RMSprop
model = Sequential()
model.add(Dense(1024, input_dim=784))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='softmax'))

# NOTE: unscaled learning rate
optimizer = keras.optimizers.SGD(lr=0.1)
loss = 'binary_crossentropy'
store = DBFSLocalStore('/dbfs/ml/exprements')
keras_estimator = hvd.KerasEstimator(
    num_proc=2,
    store=store,
    model=model,
    optimizer=optimizer,
    loss=loss,
    feature_cols=['features'],
    label_cols=['label'],
    batch_size=120,
    epochs=12)


keras_model = keras_estimator.fit(train_df).setOutputCols(['predict'])
```

```
testclass = testclass.astype('int')
```

```
pdtest = pd.DataFrame(test)
pdtestclass = pd.DataFrame(testclass, columns=['y'])
pdtestDF = pd.concat([pdtest,pdtestclass],axis=1)
testdf = spark.createDataFrame(pdtestDF)
```

```
from pyspark.ml.feature import VectorAssembler
va = VectorAssembler(inputCols=[c for c in df.columns if c!='y'], outputCol='features')
```

```
testdf = va.transform(testdf)
```

```
from pyspark.ml.feature import OneHotEncoder

ohe = OneHotEncoder(inputCols=['y'],
                            outputCols=['label'],
                            dropLast=False)
model = ohe.fit(testdf)
test_df = model.transform(testdf)
```

```
predict_df = keras_model.transform(test_df)
```
