```
from tensorflow.keras.datasets import mnist
```

```
(train,trainclass),(test,testclass) = mnist.load_data()
```

```
train = train.reshape(60000,784)
```

```
test = test.reshape(10000,784)
```

```
import pandas as pd
pd.DataFrame(train)
```

```
train = train.astype('float32') / 255
```

```
test = test.astype('float32') / 255
```

```
testclass = testclass.astype('float32')
```

```
trainclass = trainclass.astype('float32')
```

```
import pandas as pd
import numpy as np
pdtrain = pd.DataFrame(train)
```

```
pdtest = pd.DataFrame(test)
```

```
pdtrainclass = pd.DataFrame(trainclass, columns=['y'])
```

```
pdtestclass = pd.DataFrame(testclass, columns=['label'])
```

```
pdtrainclass
```

```
pdDF = pd.concat([pdtrain,pdtrainclass], axis=1)
```

```
pdDFtest = pd.concat([pdtest,pdtestclass], axis=1)
```

```
pdDFtest
```

```
df = spark.createDataFrame(pdDF)
```

```
dftest = spark.createDataFrame(pdDFtest)
```

```
dftest.printSchema()
```

```
from pyspark.ml.feature import VectorAssembler, OneHotEncoder
```

```
va = VectorAssembler(inputCols=[c for c in df.columns if c!='y'], outputCol='features')
```

```
va1 = VectorAssembler(inputCols=[c for c in dftest.columns if c!='label'], outputCol='features')
```

```
df = va.transform(df)
```

```
dftest = va1.transform(dftest)
```

```
ohe = OneHotEncoder(inputCol='y', outputCol='label', dropLast=False)
```

```
df = ohe.fit(df).transform(df)
```

```
df.rdd.getNumPartitions()
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import categorical_crossentropy

model = Sequential()
model.add(Dense(1024, input_dim=784))
model.add(Dense(512, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```
import horovod.spark.keras as hvd
from horovod.spark.common.store import DBFSLocalStore
store = DBFSLocalStore('/dbfs/ml/depplearn')

optimizer = Adam(learning_rate=0.1)
loss = categorical_crossentropy

keras_estimator = hvd.KerasEstimator(
    num_proc=8,
    store=store,
    model=model,
    optimizer=optimizer,
    loss=loss,
    feature_cols=['features'],
    label_cols=['label'],
    batch_size=128,
    epochs=5)
```

```
keras_model = keras_estimator.fit(df).setOutputCols(['predict'])
```

```
predict_df = keras_model.transform(dftest)
```

```
newdf = predict_df.toPandas()
```

```
newdf
```

```
from pyspark.sql.functions import udf
from pyspark.sql.types import DoubleType

maxVal = udf(lambda x: float(np.argmax(x)), returnType=DoubleType())
```

```
finalDF = predict_df.withColumn('prictedLabel', maxVal(predict_df.predict))
```

```
finalDF.show(vertical=True)
```

```
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

eval = MulticlassClassificationEvaluator(labelCol='label', predictionCol='prictedLabel', metricName='accuracy')
```

```
result = eval.evaluate(finalDF)
```
