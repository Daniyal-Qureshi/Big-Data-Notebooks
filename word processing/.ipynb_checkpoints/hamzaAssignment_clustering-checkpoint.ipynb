```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
#spark = sparknlp.start()
```

```python
from  pyspark.sql.functions import input_file_name

txtFiles = spark.read.text('cnn_stories/cnn/stories/*', wholetext=True)
```

```python
txtFiles.show(5,truncate=False)
```

```python
txtFiles.count()
```

```python
type(txtFiles)
txtFiles.printSchema()
```

```python
from pyspark.ml.feature import RegexTokenizer
regex = RegexTokenizer(inputCol='value', outputCol='words',pattern='\\W')
```

```python
regexDF = regex.transform(txtFiles)
```

```python
regexDF.select('words').show(5,truncate=False)
```

```python
from pyspark.ml.feature import StopWordsRemover
```

```python
dic = StopWordsRemover.loadDefaultStopWords('english')
```

```python
stopwords = StopWordsRemover(inputCol='words',stopWords=dic,outputCol='removed_words')
```

```python
stopedDF = stopwords.transform(regexDF)
```

```python
stopedDF.select('words','removed_words').show(truncate=False)
```

```python
from pyspark.ml.feature import HashingTF, IDF

tf = HashingTF(inputCol='removed_words', outputCol='rawFeatures')
```

```python
featuredDF = tf.transform(stopedDF)
```

```python
featuredDF.select('rawFeatures').show(5,truncate=False)
```

```python
idf = IDF(inputCol="rawFeatures", outputCol="features")

idf = idf.fit(featuredDF)
```

```python
idfDF = idf.transform(featuredDF)
```

```python
idfDF.select('rawFeatures', 'features').show(5,truncate=False)
```

```python
type(idfDF)
```

```python
from pyspark.ml.feature import VectorAssembler
va = VectorAssembler(inputCols=['rawFeatures', 'features'], outputCol='featuress')
vaDF = va.transform(idfDF)
vaDF.select('featuress').show(5,truncate=False)
```

```python
vaDF.show()
```

```python
#Apply PCA
```

```python
from pyspark.ml.feature import PCA
pcaDF = vaDF.select('featuress')
```

```python
pca = PCA(k=10000, inputCol="featuress", outputCol="pcaFeatures")
```

```python
pcamodel = pca.fit(pcaDF)
```

```python
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator
```

```python
eval_result =[]
evaluator = ClusteringEvaluator(featuresCol='scaledFeatures', predictionCol='prediction', metricName='silhouette', distanceMeasure='squaredEuclidean')
```

```python
kmeans = KMeans(featuresCol='features',predictionCol='prediction',k=11,initMode='k-means||',initSteps=5,tol=0.0001,
    maxIter=20,
    seed=None,
    distanceMeasure='euclidean')
    
kmeansModel = kmeans.fit(idfDF)
kmeanCluters = kmeansModel.transform(idfDF)
```

```python
kmeanCluters.rdd.getNumPartitions()
```

```python
evaResult = evaluator.evaluate(kmeanCluters)
```
