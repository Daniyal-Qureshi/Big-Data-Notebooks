```python
rdd = sc.parallelize(range(1000)).map(lambda a: (a,a)).partitionBy(10)
print(rdd.getNumPartitions())
print(rdd.glom().map(len).collect())
print(rdd.glom().collect())
print(rdd.collect())
```

```python
rdd2 = rdd.filter(lambda a: a[0]%5 == 0)
print(rdd2.glom().map(len).collect())
print(rdd2.glom().collect())
```

```python
maprdd = sc.parallelize([1, 2, 3, 4], 2)
print(maprdd.collect())
print(maprdd.glom().collect())
def f(iterator):
    yield sum(iterator)

maprdd = maprdd.mapPartitions(f)

print(maprdd.collect())
print(maprdd.glom().collect())
```

```python
ziprdd1 = sc.parallelize([('a',1),('a',2),('b',3),('b',4),('c',5),('c',6),('d',7)])
ziprdd2 = sc.parallelize([('e',9),('e',10),('f',11),('f',12),('g',13),('g',14),('g',15)])

print(ziprdd1.collect())
print("===============================")
print(ziprdd1.glom().collect())
print("===============================")

print(ziprdd2.collect())
print("===============================")
print(ziprdd2.glom().collect())
print("===============================")

ziprdd3 = ziprdd1.zip(ziprdd2)
print(ziprdd3.collect())
print("===============================")
print(ziprdd3.glom().collect())
```
